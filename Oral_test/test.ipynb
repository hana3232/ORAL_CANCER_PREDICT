```python
df=pd.read_csv("D:\\Oral_test\\oral_cancer_prediction_dataset.csv")
```

```python
df.drop(columns=["ID","Oral Cancer (Diagnosis)","Cost of Treatment (USD)","Economic Burden (Lost Workdays per Year)"],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for i in df.select_dtypes(include = ["object"]).columns:
    df[i] = le.fit_transform(df[i])
```

```python
x=df.drop("Survival Rate (5-Year, %)",axis=1)
```

```python
x
```

```python
y=df["Survival Rate (5-Year, %)"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.10)
```

```python
x_train.shape, x_test.shape, y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import AdaBoostClassifier,AdaBoostRegressor #DTC is the algorithm we are using
from sklearn.metrics import classification_report
```

```python
model= AdaBoostRegressor().fit(x_train,y_train)
y_pred=model.predict(x_test)
```
